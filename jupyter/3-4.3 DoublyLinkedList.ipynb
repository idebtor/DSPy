# Doubly Linked List and Deque

__Doubly linked list__ 의 핵심은 노드와 노드가 양방향으로 서로 연결되어 있다는 점입니다. singly linked list와는 다르게 노드가 이전 노드(previous)와 다음 노드(next)로 구성되어 있습니다. 또한 리스트의 양 끝에 `head`와 `tail`가 있어서 각각 첫 노드와 마지막 노드와 연결되어 있습니다.

위의 그림처럼, 리스트의 양쪽 끝에 `head`와 `tail` 노드들이 있고, 각 노드는 `prev`, `data`, `next`로 이루어져 있습니다. 리스트의 양 끝에 추가된 두 개의 노드 즉 `head`와 `tail`은 `sentinel` 노드라고 부릅니다. 

`Doubly Linked List`의 장점은 양방향으로 연결되어 있기 때문에 노드를 탐색이나 노드를 추가 및 삭제하는 방향이 한쪽이 아닌 양쪽으로 가능하다는 것입니다. 또한, 데이터 가진 노드들이 존재한다면, 항상 두 개의 센티널 노드들 사이에 있기 때문에서, 노드의 위치(처음, 중간 끝)에 따라 코딩을 달리해야 하는 Singly Linked List보다 코딩이 오히려 간편합니다.

## Define DoublyLinked class
이제 `Node` 클래스를 사용하여 DoublyLinked 클래스의 `__init__()` constructor(생성자 메소드)를 정의합니다. 여기서는 일단 가장 기본적인 형태로 정의합니다. 
- 여기 생성자 매소드를 이해하는 것이 중요합니다. DoublyLinked() 생성자를 사용하면 쉽게 head/tail 노드만으로 구성된 empty DoublyLinked 객체가 생성이 됩니다. 
- 메소드 `begin()`은 `__head` 노드를 가리키는 __데이터가 있는 첫번째 노드__ 를 반환합니다. 데이터가 있는 노드가 없다면 `tail`노드를 반환합니다. 
- 메소드 `end()`는 리스트의 Sentinel 노드 즉 `__tail` 노드를 반환합니다.
- `is_empty()`는 리스트의 __데이터가 있는 노드__ 가 하나라도 있으면, True를 반환합니다. Sentinel 노드는 고려하지 않습니다. 
- `is_empty()` 메소드에서 `__head`와 `__tail`를 사용하지 않은 것처럼, 특별한 경우를 제외하고, 코드의 편리성을 위하여 `begin()`과 `end()`함수 사용을 권장합니다. 그러한 의미에서 `head`, `tail`대신에 `__head`, `__tail`이란 이름을 사용합니다.

````python
class DoublyLinked:
    def __init__(self):
        self.__head = Node()
        self.__tail = Node()
        self.__head.next = self.__tail
        self.__tail.prev = self.__head
        self.__tail.next = None

    def begin(self):
        return self.__head.next
    
    def end(self):
        return self.__tail

    def is_empty(self):
        return self.begin() == self.end()
        
if __name__ == '__main__':
    # create a DoublyLinked object and name it as 'alist' 
    alist = None
    print('begin():', None)           # replace None with alist...
    print('  end():', None             # replace None with alist...
    print('empty():', None)        # replace None with alist...
````

## Define Key operations: `remove()` and `insert()`

가장 핵심적인 두 메소드를 정의합니다. 

- `remove(x)`는 지정된 단일 노드 `x`를 리스트에서 제거합니다. 'x'를 발견할 수 없으면, 'None'을 반환하고, 성공적으로 제거하면 값을 반환합니다. 
- `insert(data, x)`는 `data`를 가진 새로운 노드를 기존하는 노드 `x` 위치에 삽입합니다. 즉 노드 `x`와 그 이후 노드는 하나씩 오른쪽으로 밀려 납니다. 

- 이 두 메소드를 사용하여, 다른 작업들을 쉽게 코딩할 수 있습니다.

```
    - self.remove(self.end().prev)        # pop() - remove the last node
    - self.remove(self.begin())           # popleft() - remove the first node
    - self.insert(data, self.end())       # append(data) - insert a new node at the end
    - self.insert(data, self.begin())     # appendleft(data) - insert a new node at the front
```    

- 노드 `x`가 리스트에 존재하기만 하면, 그의 위치에 관계없이 O(1) 시간 복잡도로 삽입 및 제거할 수 있습니다.
