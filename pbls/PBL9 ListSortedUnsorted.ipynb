# Linked List
- 노드 클래스와 링크드 리스트의 코드는 다음과 같습니다.
- 아래 구현된 코드에 대한 설명은 `Ch3-3.1 LinkedList.ipynb` 파일을 참고하십시오.
- 가장 먼저 리스트를 출력할 수 있게 하는 코드를 개발하고, 
- 다른 유용한 메소드들과 `Discursison Questions`에서 언급한 `pop(item)` 메소드의 관련한 사항들을 다루어 봅니다.

````python
class Node:
    def __init__(self,initdata):
        self._data = initdata
        self._next = None

    def get_data(self):
        return self._data

    def get_next(self):
        return self._next

    def set_data(self,newdata):
        self._data = newdata

    def set_next(self,newnext):
        self._next = newnext        
````

````python
class ListUnsorted:

    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None

    def push(self,item):
        temp = Node(item)
        temp.set_next(self.head)
        self.head = temp

    def size(self):
        current = self.head
        count = 0
        while current != None:
            count = count + 1
            current = current.get_next()

        return count

    def find(self,item):
        current = self.head
        found = False
        while current != None and not found:
            if current.get_data() == item:
                found = True
            else:
                current = current.get_next()

        return found

    def pop(self,item):
        current = self.head
        previous = None
        found = False
        while not found:
            if current.get_data() == item:
                found = True
            else:
                previous = current
                current = current.get_next()

        if previous == None:
            self.head = current.get_next()
        else:
            previous.set_next(current.get_next())
````

````python
if __name__ == '__main__':
    mylist = ListUnsorted()

    mylist.push(31)
    mylist.push(77)
    mylist.push(17)
    mylist.push(93)
    mylist.push(26)
    mylist.push(54)

    print(mylist.size())       # 6
    print(mylist.find(93))   # True
    print(mylist.find(100))  # False

    mylist.push(100)
    print(mylist.find(100))  # True
    print(mylist.size())       # 7

    mylist.pop(54)
    print(mylist.size())    # 6
    mylist.pop(93)
    print(mylist.size())    # 5
    mylist.pop(31)
    print(mylist.size())    # 4
    print(mylist.find(93))  # False

    print(mylist)
````
